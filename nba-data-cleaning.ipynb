### Data Cleaning of NBA Data

```python
import pandas as pd
```

```python
mvps = pd.read_csv("mvps.csv")
```

```python
mvps
```

```python
# Subset of mvps with fewer columns

mvps = mvps[['Player', 'year','Pts Won','Pts Max','Share']]
```

```python
mvps
```

```python
players = pd.read_csv("players.csv")
```

```python
players
```

```python
# deleteing first two columns

# del players["Rk"]
del players['Unnamed: 0']
```

```python
players.head(100)
```

```python
# cleaning players name column - removing unwanted charcters

players['Player'] = players['Player'].str.replace("*","", regex=False)
```

```python
players.head()
```

grouping players data into one row if same player has more than two rows.


```python
players.groupby(['Player','year'])
```

```python
# working on Gerg andersone data, and merging all rows for him in one

def single_row(df):
    if df.shape[0] == 1:
        return df
    else:
        row = df[df['Tm'] == 'TOT']
        row['Tm'] = df.iloc[-1,:]['Tm']
        return row 
    
```

```python
players = players.groupby(['Player','year']).apply(single_row)
```

```python
players
```

```python
players.index = players.index.droplevel()
```

```python
players.index = players.index.droplevel()
```

```python
players[players['Player'] == "Greg Anderson"]
```

```python
combined_df = players.merge(mvps, how='outer', on=['Player','year'])
```

```python
combined_df
```

```python
combined_df[combined_df['Pts Won'] > 0]
```

```python
combined_df[['Pts Won', 'Pts Max','Share']] = combined_df[['Pts Won', 'Pts Max','Share']].fillna(0)
```

```python
combined_df[combined_df['Pts Won'] > 0]
```

```python
combined_df[combined_df['Player'] == 'Kobe Bryant']
```

### Team data CSV

```python
teams = pd.read_csv("teams.csv")
```

```python
teams.head(10)
```

```python
# remvoing text row from data

teams = teams[~teams['W'].str.contains('Division')]
```

```python
teams.head(20)
```

```python
# removong unwanted charcters

teams['Team'] = teams['Team'].str.replace("*","", regex=False)
```

```python
teams.head(5)
```

#### teams dataset has full name of teams while combined_df has three characters

```python
teams['Team'].unique()
```

```python
combined_df['Tm'].unique()
```

```python
# reading nicknames.txt file

nicknames = {}

with open("nicknames.txt") as f:
    lines = f.readlines()
    for line in lines[1:]:
        abbrev, name = line.replace("\n", "").split(",")
        nicknames[abbrev] = name
```

```python
combined_df['Team'] = combined_df['Tm'].map(nicknames)
```

```python
combined_df.head()
```

```python
stats = combined_df.merge(teams, how="outer", on=['Team','year'])
```

```python
stats
```

```python
del stats['Unnamed: 0']
```

```python
# data types
stats.dtypes
```

```python
# converting datatypes to numeric  as most are objects

stats = stats.apply(pd.to_numeric, errors='ignore')
```

```python
stats['GB'].unique()

# GB column has '-' value. Repalcing it with 0
```

```python
stats['GB'] = stats['GB'].str.replace('—', "0")
```

```python
stats['GB'].unique()
```

```python
# converting GB datatype to numeric 

stats['GB'] = pd.to_numeric(stats['GB'])
```

```python
stats.dtypes
```

### writng complete data to csv
