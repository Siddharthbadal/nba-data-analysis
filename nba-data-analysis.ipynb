```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
stats = pd.read_csv("player_mvp_stats.csv")
```

```python
stats
```

```python
# deleting the 
del stats['Unnamed: 0']
```

### Handling Missing Values

```python
pd.isnull(stats).sum()
```

```python
# Michael Jordan

stats[stats['Player']=='Michael Jordan']
```

### Top Scores

```python
top_scores = stats[stats['G']>75].sort_values("PTS", ascending=False).head(10)
top_scores
```

```python
top_scores.plot.bar("Player","PTS")
```

### Best player for each year

```python
yearly_best_player = stats.groupby('year').apply(lambda x: x.sort_values("PTS", ascending=False).head(1))
```

```python
yearly_best_player.head()
```

```python
stats = stats.fillna(0)
```

```python
stats.shape
```

```python
# changing column name 'year' to 'Year'

stats = stats.rename({'year':'Year'}, axis='columns')
```

```python
stats.columns
```

```python
# number of players of each Team

stats['Team'].value_counts()
```

```python
plt.figure(figsize=(15,5))
plt.bar(list(stats['Team'].value_counts()[0:10].keys()), list(stats['Team'].value_counts()[0:10]),color='g')
plt.xticks(rotation=60)
plt.show()
```

```python
# grouping data by Teams and finding values
stats.groupby('Team')['Pts Won'].agg(['count','sum','max','mean'])
```

```python
# Win-Lose Percentage of Teams
stats.groupby('Team')['W/L%'].max()
```

```python
stats.groupby('Player')['W/L%'].max().sort_values(ascending=False)[:25]
```

```python
# Total wins by each team

stats.groupby('Team')['W'].max()
```
